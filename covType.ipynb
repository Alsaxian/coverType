Let's do some simple PCA first to visualize the cover sub-groups.

```python
from sklearn.decomposition import PCA
pca_unreweighted = PCA(n_components = 2)
X_train_scores_unreweighted = pca_unreweighted.fit(X_train_scaled).transform(X_train_scaled)
pca_unreweighted.explained_variance_ratio_

plt.figure(figsize=(20,6))
colors = sns.color_palette("Set2", 7)
lw = 2
Y_train_to_colors = Y_train.map(lambda x: colors[x-1])

plt.scatter(X_train_scores_unreweighted[:, 0], X_train_scores_unreweighted[:, 1], color=Y_train_to_colors, alpha=.8, lw=lw, label=cover_type)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of cover type')
plt.show()

pca_reweighted = PCA(n_components = 2)
X_train_reweighted = X_train_scaled
X_train_reweighted[:, 10:14] = X_train_reweighted[:, 10:14] / np.sqrt(4)
X_train_reweighted[:, 14:52] = X_train_reweighted[:, 14:54] / np.sqrt(40)
X_train_scores_reweighted = pca_reweighted.fit(X_train_reweighted).transform(X_train_reweighted)
pca_reweighted.explained_variance_ratio_

plt.figure(figsize=(20,6))
plt.scatter(X_train_scores_reweighted[:, 0], X_train_scores_reweighted[:, 1], color=Y_train_to_colors, alpha=.8, lw=lw, label=cover_type)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of cover type reweighted X')
plt.show()
```

```python
pca_numeric_only = PCA(n_components = 2)
X_train_scores_numeric_only = pca_numeric_only.fit(X_train_scaled[:, 0:10]).transform(X_train_scaled[:, 0:10])
pca_numeric_only.explained_variance_ratio_

plt.figure(figsize=(20,6))
colors = sns.color_palette("Set2", 7)
lw = 2
Y_train_to_colors = Y_train.map(lambda x: colors[x-1])

plt.scatter(X_train_scores_numeric_only[:, 0], X_train_scores_numeric_only[:, 1], color=Y_train_to_colors, alpha=.8, lw=lw, label=cover_type)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of cover type 3')
plt.show()

pca_catagorical_only = PCA(n_components = 2)
X_train_scores_catagorical_only = pca_catagorical_only.fit(X_train_scaled[:, 10:52]).transform(X_train_scaled[:, 10:52])
pca_catagorical_only.explained_variance_ratio_

plt.figure(figsize=(20,6))

plt.scatter(X_train_scores_catagorical_only[:, 0], X_train_scores_catagorical_only[:, 1], color=Y_train_to_colors, alpha=.8, lw=lw, label=cover_type)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of cover type 4')
plt.show()
```

```python
pca_reweighted_deeper = PCA(n_components = 2)
X_train_reweighted_deeper = X_train_scaled
X_train_reweighted_deeper[:, 10:14] = X_train_reweighted_deeper[:, 10:14] / 4
X_train_reweighted_deeper[:, 14:54] = X_train_reweighted[:, 14:54] / 40
X_train_scores_reweighted_deeper = pca_reweighted_deeper.fit(X_train_reweighted_deeper).transform(X_train_reweighted_deeper)
pca_reweighted_deeper.explained_variance_ratio_

plt.figure(figsize=(20,6))
plt.scatter(X_train_scores_reweighted_deeper[:, 0], X_train_scores_reweighted_deeper[:, 1], color=Y_train_to_colors, alpha=.8, lw=lw, label=cover_type)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of cover type reweighted deeper X')
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

### ordinary LDA
lda_ordinaire = LinearDiscriminantAnalysis()
lda_ordinaire.fit(X_train, Y_train)
# lda_ordinaire.predict(X_test)
lda_ordinaire.score(X_test, Y_test)
```

```python
### Auto-schrinked LDA

lda_shrink_auto = LinearDiscriminantAnalysis(solver = "lsqr", shrinkage = "auto")
lda_shrink_auto.fit(X_train, Y_train)
lda_shrink_auto.score(X_test, Y_test)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda_ordinaire = QuadraticDiscriminantAnalysis()
qda_ordinaire.fit(X_train, Y_train)
qda_ordinaire.score(X_test, Y_test)
```

```python
### Shrinkage with grid, method least square
lda_shrink_manuel = pd.Series(np.arange(0.1, 1, 0.1)).map(lambda i: LinearDiscriminantAnalysis(solver = "lsqr", shrinkage = i))
lda_shrink_manuel[1]
lda_shrink_manuel.map(lambda model: model.fit(X_train, Y_train))
lda_shrink_manuel.map(lambda model: model.score(X_test, Y_test))
```

```python
### Shrinkage with grid, method eigen-decomposition
lda_shrink_manuel = pd.Series(np.arange(0.1, 1, 0.1)).map(lambda i: LinearDiscriminantAnalysis(solver = "eigen", shrinkage = i))
lda_shrink_manuel[1]
lda_shrink_manuel.map(lambda model: model.fit(X_train, Y_train))
lda_shrink_manuel.map(lambda model: model.score(X_test, Y_test))
```

```python
lda_shrink_reduction = pd.Series(np.arange(1, 6)).map(lambda i: LinearDiscriminantAnalysis(n_components = i))
lda_shrink_reduction.map(lambda model: model.fit(X_train, Y_train))
lda_shrink_reduction.map(lambda model: model.score(X_test, Y_test))
```
